## Putting data on a map

```python
import matplotlib.pylab as plt
import xarray as xr
import numpy as np
from matplotlib import cm
# xr.set_options(display_style="text") #uncomment if you can't display html 
```

Let's have a look at our Reanalysis data once again:

```python
ds = xr.open_dataset('../data/air.2m.gauss.2021.nc')
```

```python
ds
```

We can have a look at the data

```python
plt.imshow(ds.air[0,:,:])
plt.colorbar()
```

But our lons and lats should span the complete range global range:

```python
ds.lon
```

```python
ds.lat
```

You can provide coordinates to python functions, of cours, and get something like this:

Prepear our data (just make separate variables):

```python
lon = ds.lon.data
lat = ds.lat.data
temp = ds.air[0,:,:].data - 273.15 # convert to Celsius
```

```python
plt.contourf(lon, lat, temp, levels=np.linspace(-40, 40, 41), cmap=cm.RdBu_r)
plt.colorbar()
```

Looks better but not really as a map. We want to use different projections, put coastlines and so on. For this we will use [cartopy](https://scitools.org.uk/cartopy/docs/latest/#)

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

Simplest map is an empty one:

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())
```

add coastlines

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', color='k')
```

We can change progection ([list of available projections](https://scitools.org.uk/cartopy/docs/latest/crs/projections.html)).

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines(resolution='110m', color='k')
```

Ok, let's add our data to the map:

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines(resolution='110m', color='k')
ax.contourf(lon, lat, 
            temp, levels=np.linspace(-40, 40, 41), 
            cmap=cm.RdBu_r, transform=ccrs.PlateCarree())
```

There is an ugly gap in the middle due to the ways our lons are organised:

```python
lon
```

```python
from cartopy.util import add_cyclic_point
```

```python
temp_cyc, lon_cyc = add_cyclic_point(temp, coord=lon)
```

```python
lon_cyc
```

```python
temp.shape
```

```python
temp_cyc.shape
```

```python
temp[:,0]-temp[:,-1]
```

```python
temp_cyc[:,0]-temp_cyc[:,-1]
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.Robinson())
ax.coastlines(resolution='110m', color='k')
ax.contourf(lon_cyc, lat, 
            temp_cyc, levels=np.linspace(-40, 40, 41), 
            cmap=cm.RdBu_r, transform=ccrs.PlateCarree())
```

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines(resolution='110m', color='k')
ax.contourf(lon_cyc, lat, 
            temp_cyc, levels=np.linspace(-40, 40, 41), 
            cmap=cm.RdBu_r, transform=ccrs.PlateCarree())
```

For some projections you have to specify the extent, otherwise they will not work:

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.NorthPolarStereo())
ax.coastlines(resolution='110m', color='k')
ax.set_extent((-180, 180, 60, 90),crs=ccrs.PlateCarree())
ax.contourf(lon_cyc, lat, 
            temp_cyc, levels=np.linspace(-40, 40, 41), 
            cmap=cm.RdBu_r, transform=ccrs.PlateCarree())
```

You can specufy extent for other projections as well:

```python
plt.figure(figsize=(10,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines(resolution='10m', color='k')

ax.set_extent((5, 35, 50, 67), crs=ccrs.PlateCarree())

img = ax.contourf(lon_cyc, lat, 
            temp_cyc, levels=np.linspace(-20, 20, 41), 
            cmap=cm.RdBu_r, transform=ccrs.PlateCarree())
plt.colorbar(img)
```
